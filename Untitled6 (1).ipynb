```python
import matplotlib.pyplot as plt
```

```python
def open_file(path):
  data = []
  with open(path) as file:
   for i in file.readlines()[1:]:
    data.append(i.strip("\n").split(","))
  return data
```

```python
data = open_file("./movies_data_with_budget.csv")
```

```python
print(data)
```

```python
dic = {}

for movie in data:
    # print(movie[2])
    dic[movie[2]] = dic.get(movie[2],0) + 1

rate = {}

for i in dic.keys():
    rate[i] = rate.get(i,0) + 1

for i in data:
    rate[i[2]] = rate[i[2]] + float(i[1])

for i in dic.keys():
    rate[i] = rate[i]/dic[i]

print(rate
      )

# print(dic)
# plt.hist([i for i in dic.keys()], [i for i in dic.values()],width=0.6,bins=50)
# plt.figure(figsize=(300,300))
# plt.xlabel="Genres"
# plt.ylabel="number of movie"
# # plt.bar_label(True)
# plt.show()




fig, ax = plt.subplots()
ax.bar([i for i in dic.keys()], [i for i in dic.values()], color='skyblue', edgecolor='black')

# Adding average rating as text above each bar
for i, val in dic.items():
    plt.text(i,val,f"{rate[i]:.2f}", ha='center', fontsize=10)

# Title and labels
plt.title('Number of Movies vs Genre (with Average Rating)', fontsize=14)
# plt.xlabel('Genre')
# plt.ylabel('Number of Movies')
ax.set_xlabel('Performance')
# Rotating x-axis labels for better readability
plt.xticks(rotation=45)

# Display the plot
plt.tight_layout()
plt.show()
```
